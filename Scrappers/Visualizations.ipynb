```python
import pandas as pd
import requests, json
from config import fred_api_key, quandl_api_key, eia_api_key
import datetime as dt
import pprint
import os
import matplotlib.pyplot as plt
```

```python
combined_df = pd.read_csv(filepath_or_buffer = '../data/combined_df.csv')
combined_df.head()
```

```python
economic_indicators_df = pd.read_csv(filepath_or_buffer = '../data/EconomicIndicators.csv')
economic_indicators_df.head()
```

```python
for index, row in combined_df.iterrows():
    combined_df['date'][index] = dt.date.fromisoformat(row.date)
```

```python
for index, row in economic_indicators_df[0:5].iterrows():
    economic_indicators_df.Date[index] = dt.date.fromisoformat(row.Date)
```

```python
plt.figure(num=1, figsize=(30,10))
plt.plot(combined_df.date, combined_df.Contraction)
plt.title('Binary Recession by Quarter')
plt.savefig('../Images/Binary Recession by Quarter.png')
plt.show()
```

```python
plt.figure(num=2, figsize=(30,10))
plt.plot(combined_df.date, combined_df.consumer_sentiment)
plt.title('Consumer Sentiment')
plt.savefig('../Images/Consumer_Sentiment.png')
plt.show()
```

```python
plt.figure(num=3, figsize=(30,10))
plt.plot(economic_indicators_df.Date, economic_indicators_df['GDP pct change'])
plt.title('GDP pct change')
plt.savefig('../Images/GDP_pct_change.png')
plt.show()
```

```python
plt.figure(num=4, figsize=(30,10))
plt.plot(combined_df.date, combined_df.deficit)
plt.title('Deficit')
plt.savefig('../Images/Deficit.png')
plt.show()
```

```python
plt.figure(num=5, figsize=(30,10))
plt.plot(combined_df.date, combined_df.copper_price)
plt.title('Copper Price')
plt.savefig('../Images/Copper_Price.png')
plt.show()
```
